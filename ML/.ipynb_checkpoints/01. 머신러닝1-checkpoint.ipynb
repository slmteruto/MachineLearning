### 일반화, 과대적합(Overfitting), 과소적합(Underfitting)

    할 때 훈련용, 테스트용(완전 새로운 데이터로 블라인드해야함), 실전용으로 나눈다.     
    훈련을 시키고 정확도가 90% 이상 넘어가기 시작하면 테스트    
    근데 테스트용을 넣었는데 정확도가 20%..?   왜? -> 훈련데이터가 너무 복잡하게 만들어져서 그런것
    훈련에만 최적화가 되어있음. -> 이게 '과대적합'이다.
    
    해결방법은?  훈련용데이터를 너무 정확히 하려고 하지 않아야 한다. 몇개 틀린것도 허용하고
    이렇게 맞춰가는 과정을 '일반화' 라고 함
    
    '과소적합'은 너어어무 단순하게 만들어서 결과가 의미없게 나와버리는것
    
+ 과대적합 : 훈련세트의 각 샘플에 너무 가깝게 맞춰져 있어서 복잡한 모델을 만드는 것
+ 과소적합 : 너무 간단한 모델이 선택되는 것
+ 일반화 : 처음보는 데이터에 대해서 정확하게 예측할 수 있는것

### L1 norm, L2 norm


```python
# L1 Norm
# 주사위 게임(-3 ~ 3) - 숫자는 이동

# 3은세칸이동, -3은 뒤로 세칸이동

# 세명이 세번을 던질 때 누가 제일 많이 이동했냐를 구할 때 어떻게 푸는것이 가장 효율적인가
# 앞뒤 상관없이 젤 많이 걸은것

df = pd.DataFrame(np.random.randint(low=-3, high=3, size=(3,3))) # 최소, 최대, 3행3열로
df.columns = ["move_1", "move_2","move_3"]  
df.head()

# l1 norm일지 l2 norm 선택할지 선택하는 함수가 있다.
from numpy import linalg

# 누가 많이 걸었을까
movements = df.values
movements

linalg.norm(movements, ord=1, axis=1) #ord : 순서,  축(axis)은 행방향 
# 두번째가 젤 많이 걸었다. 
```

```python
# L2 Norm

# 2차원 좌표에서 거리구할 때 주로 사용

df = pd.DataFrame(np.random.randint(low=1, high=10, size=(3,2)))
df.columns= ["x", "y"]
df

sns.lmplot('x', 'y', data=df, fit_reg=False, scatter_kws={"s":200})

# L2 norm 거리를 구하기 위한
data_points = df.values
data_points

linalg.norm(data_points, ord=2, axis=1)
# 결과가 매번 달라지긴 하는데 0에서 가장 가까운건 젤 숫자가 작은 값이다.
```

 ### KNN 구현

```python
X, y = mglearn.datasets.make_forge()
# !pip install sklearn
#테스트, 훈련용 나눠야함

#scikit-learn
import mglearn
from sklearn.model_selection import train_test_split  #이걸 사용하면 좀 더 쉽게 나눌 수 있다.


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)  #0 :랜덤 값 고정
# X : 입력값, y : 결과값
# 모든 데이터를 이렇게 나누는게 좋다. 

print(X_train.shape, X_test.shape)  
# 28개중에 19개를 훈련데이터로 만든것
# 28개중에 7개 테스트값

print(y_train, y_test)  
# 28개중에 19개를 훈련 결과값
# 28개중에 7개 테스트 결과값
```

```python
# 교육시키기 : 지도방식
from sklearn.neighbors import KNeighborsClassifier  
clf= KNeighborsClassifier(n_neighbors=3)   # 이웃의 갯수가 너무많아도 문제 적어도 문제, 적으면 모든걸 다 확인
clf.fit(X_train, y_train)
```

```python
# 학습(훈련) 결과 테스트
clf.predict(X_test)  #테스트데이터를 넘겨주면 예측을 해준다

# 1 0 1 0 1 0 0 으로 예측했다.
```

```python
# 실제정답과 비교
clf.score(X_test, y_test)  #정확도를 측정하는 함수. 테스트값과 테스트결과값을 넘겨준다.
# 0.86 , 86%의 정확도를 보인다. 
```

```python
#이웃의 갯수 1개일때
clf = KNeighborsClassifier(n_neighbors=8)
clf.fit(X_train, y_train) 

# 예측도는 예측을 먼저 적나 훈련을 먼저하나 똑같음
# 훈련 세트의 정확도
print(clf.score(X_train, y_train)) 

#실제 테스트 정확도
print(clf.score(X_test, y_test))


# 훈련은 100% 다 맞추었지만
# 결과는 90%로 했다.

# 갯수가 8일때가 젤 높긴하다.
```

```python
train_acc = []
test_acc = []

for n_neigh in range(1, 11) :
    clf = KNeighborsClassifier(n_neighbors=n_neigh)
    clf.fit(X_train, y_train)
    
    #훈련 세트의 정확도
    train_acc.append(clf.score(X_train, y_train))
    
    #테스트 세트의 정확도
    test_acc.append(clf.score(X_test, y_test))
    
plt.plot(range(1, 11), train_acc, label="훈련정확도")
plt.plot(range(1, 11), test_acc, label="테스트정확도")

#결과를 보면 테스트가 6개일때가 가장 결과가 높넹 

plt.ylabel("정확도")
plt.xlabel("이웃의 갯수")
plt.legend()

print(test_acc[1])  #이웃 2개 (두번째니까 인덱스 1)  -> 0.88은 나온다.
```

## 영화 분류 프로그램 만들기

```python
# 영화 샘플 
import operator

def createDataSet() :
    group = np.array([[1.0, 2.0], [1.0, 4.0], [4.0, 1.0], [4.0, 2.0]])
    label = ["Action", "Action", "Romance", "Romance"]
    return group, label

#calc 각 거리 계산하는 것 만들기
# 여기 들어가는 인자 data는 X값의 첫번째이다. 
def calcDistance(x_new, data) :
    # data 개수
    dataSize = data.shape[0] # 4개로 분류했으니 4로 나올것이다.
    
    #각 비교하는 함수를 만들기는 귀찮으니까 tile을 쓴다. (붙이는 느낌)
    #tile 정리는 밑에
    # 1차원 배열, 반복횟수는 (dataSize, 1) 즉 (4, 1) = 4행1열이라는 뜻
    diffMat = np.tile(x_new, (dataSize, 1)) - data
    
    #L2 norm
    sqDiffMat = diffMat**2
    sqDistance = sqDiffMat.sum(axis=1)
    distance = sqDistance**0.5
    sortedDistance = distance.argsort() #인덱스로 위치를 찾는함수
    print(sortedDistance) #첫번째 두번째가 제일 가깝
    return sortedDistance
    

def findMajorityClass(x_new, target, k, sortedDistance) :
    # Action, Romance를 키값으로 해서 카운트 개수를 value로 한다.
    # 예시 {"Action":8, "Romance":2}
    classCount = {}
    
    #이웃의 수만큼 반복을 돌릴 것이다. 
    for i in range(k) :
        #label이름 뽑기 (Action, Romance)
        voteLabel = target[sortedDistance[i]]
        classCount[voteLabel] = classCount.get(voteLabel, 0) + 1
        print(classCount)
        
    # 밑에 dict sort에 대해 정리하였음.
    return sorted(classCount.items(),key=operator.itemgetter(1), reverse=True)


#우리가 파이널 프로젝트가 해야하는 것 
#knn 수작업으로 만들어보자 
def classify(x_new, data, target, k) :
    sortedDistance = calcDistance(x_new, data) # 결과값이 그래서 0,1,3,2
    
    sortedDistanceCount = findMajorityClass(x_new, target, k, sortedDistance)
    #데이터, 정답, 이웃갯수, 가까운 순서
    print(sortedDistanceCount)
    
    return sortedDistanceCount
    
```

```python
X, y = createDataSet()
result = classify([2.0,3.0], X, y, 3)
#위에서 첫번째 값의 첫번째를 뽑아야 장르가 뽑힌다. 
print("결과는 {} 영화다".format(result[0][0]))
```

```python
#tile
A = 1  # 스칼라
B = np.array([0, 1]) # 1차원배열
C = np.array([[0, 1], [2, 3]]) #2차원 배열

# 위와같이 세가지 데이터 타입이 있다. 
# 결과를 보면
np.tile(A, 3)  # 결과 : array([1, 1, 1])
np.tile(B, 3)  # 결과 : array([0, 1, 0, 1, 0, 1])
np.tile(C, 3)  # 결과 : array([[0, 1, 0, 1, 0, 1],
                              #[2, 3, 2, 3, 2, 3]])

#뒤의 인자를 배열로 줄수도 있다. 
np.tile(A, (2,3)) # 결과 : array([[1, 1, 1],
                                #[1, 1, 1]])
np.tile(C, (2,3)) # 결과 : array([[0, 1, 0, 1, 0, 1],
                                 #[2, 3, 2, 3, 2, 3],
                                 #[0, 1, 0, 1, 0, 1],
                                 #[2, 3, 2, 3, 2, 3]])
```
