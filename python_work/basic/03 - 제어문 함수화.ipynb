```python
# 1 (완료)
# 가위바위보 업그레이드
# 반복문
# 결과에 승패 출력
# 종료는 x나 q

#사용자가 '가위'일때
def sciss_func(cpt) :
    global lose_cnt
    global win_cnt
    global same_cnt
    
    if cpt == 2 :
        print("졌지만 잘 싸웠습니다.")
        lose_cnt += 1
        print("*"*20)
        print("%d승, %d패, %d무"%(win_cnt, lose_cnt, same_cnt))
        print("*"*20)
    elif cpt == 3 :
        print("승리")
        win_cnt += 1
        print("*"*20)
        print("%d승, %d패, %d무"%(win_cnt, lose_cnt, same_cnt))
        print("*"*20)
    else :
        print("비겼습니다.")
        same_cnt +=1
        print("*"*20)
        print("%d승, %d패, %d무"%(win_cnt, lose_cnt, same_cnt))
        print("*"*20)
    return 

#사용자가 '바위'일때
def rock_func(cpt) :
    global lose_cnt
    global win_cnt
    global same_cnt
    
    if cpt == 3 :
        print("졌지만 잘 싸웠습니다.")
        lose_cnt += 1
        print("*"*20)
        print("%d승, %d패, %d무"%(win_cnt, lose_cnt, same_cnt))
        print("*"*20)
    elif cpt == 1 :
        print("승리")
        win_cnt += 1
        print("*"*20)
        print("%d승, %d패, %d무"%(win_cnt, lose_cnt, same_cnt))
        print("*"*20)
    else :
        print("비겼습니다.")
        same_cnt +=1
        print("*"*20)
        print("%d승, %d패, %d무"%(win_cnt, lose_cnt, same_cnt))
        print("*"*20)
    return
    

#사용자가 '보'일때
def paper_func(cpt) :
    global lose_cnt
    global win_cnt
    global same_cnt
    
    if cpt == 1 :
        print("졌지만 잘 싸웠습니다.")
        lose_cnt += 1
        print("*"*20)
        print("%d승, %d패, %d무"%(win_cnt, lose_cnt, same_cnt))
        print("*"*20)
    elif cpt == 2 :
        print("승리")
        win_cnt += 1
        print("*"*20)
        print("%d승, %d패, %d무"%(win_cnt, lose_cnt, same_cnt))
        print("*"*20)
    else :
        print("비겼습니다.")
        same_cnt +=1
        print("*"*20)
        print("%d승, %d패, %d무"%(win_cnt, lose_cnt, same_cnt))
        print("*"*20)
    
    return
########################################################################

import random

#가위 = 1
#바위 = 2
#보   = 3

win_cnt = 0    #이긴 횟수
lose_cnt = 0   # 진 횟수
same_cnt = 0   #비긴 횟수


while True :  
    user = ""                                 ## 유저가 낸 것    
    cpt = random.randint(1,3)                 ## 컴퓨터가 낸 것

    
    user = input("무엇을 내시겠습니까(가위, 바위, 보 중), (종료는 x나 q를 입력해주세요) : ")   
    
    if user == "가위" :
        print("사용자는 %s"%user)
        user = int(1)
    elif user == "바위" :
        print("사용자는 %s"%user)
        user = int(2)
    elif user == "보" :
        print("사용자는 %s"%user)
        user = int(3)
    elif user == "x" or user == "q" :
        print("종료")
        break
    else : 
        print("다시 입력하세요")
    

    if cpt == 1 :
        print("컴퓨터는 가위")
    elif cpt == 2 :
        print("컴퓨터는 바위")        
    elif cpt == 3 :
        print("컴퓨터는 보")        

    #가위바위보 결과
    if user == 1 :
        sciss_func(cpt)
    elif user == 2 :
        rock_func(cpt)
    elif user == 3 :
        paper_func(cpt)
```

```python
# 2(완료)
#간단한 계산기(4칙연산만)
# 1) 연산자 입력  예) +
# 2) 숫자1 입력 
# 3) 숫자2 입력
# 4) 결과 : 숫자1 + 숫자2 = 결과
# 5) for문 추가해서 계속 사용 가능하게


def sum_func(num1, num2) :
    result = num1+num2
    print("%d + %d = %d 입니다."%(num1, num2, result))
    return result

def min_func(num1, num2) :
    result = num1-num2
    print("%d - %d = %d 입니다."%(num1, num2, result))
    return result

def mul_func(num1, num2) :
    result = num1*num2
    print("%d * %d = %d 입니다."%(num1, num2, result))
    return result

def div_func(num1, num2) :
    result = num1/num2
    print("%d / %d = %d 입니다."%(num1, num2, result))
    return result


###################################################################

while True :
    result = 0
    func = ""
    funclist = ["+", "-", "*", "/"]
    num1 = 0
    num2 = 0
    
    func = input("연산자 입력 (+,-,*,/), x를 누르면 종료됩니다.")
    if func == "x" or func == "X" :
        break
    elif func not in funclist :
        print("잘못 누르셨습니다.")
        continue
        
    
    num1 = int(input("숫자입력1 : "))
    if func == "/" :
        num2 = int(input("숫자입력2 (0불가) : "))
        if num2 == 0 :
            print("0으로 나누는 계산은 불가능합니다.")
            continue
    else :
        num2 = int(input("숫자입력2 : "))

    if func == "+" :
        sum_func(num1, num2)
    elif func == "-" :
        min_func(num1, num2)
    elif func == "*" :
        mul_func(num1, num2)
    elif func == "/" :
        div_func(num1, num2)
    else :
        break;
```

```python
# 3(완료)
# 숫자를 입력 받아서 짝수의 합과 홀수의 합을 구하라
# 0입력하면 종료

def odd_even(num):
    print(num)
    odd_num = 0
    even_num = 0
    arr = list(range(num))  
    for i in range(num) :
        arr[i] = i+1    
    print(arr)
    
    for i in arr :
        if i % 2 != 0 :
            odd_num += i
        else :
            even_num += i    
    return odd_num, even_num

####################################################

while True :
    num = 0
    i = 0
    cnt = 0

    num = int(input("숫자 입력(0은 종료) : "))
    if num == 0 :
        print("종료")
        break
    else :
        result = odd_even(num)
    print("*"*20)
    print("홀수의 합은 %d 입니다."%result[0])
    print("짝수의 합은 %d 입니다."%result[1])
    print("*"*20)
```

```python
# 5(완료)
# 팩토리얼 계산  
# 숫자 입력하면  x!

# x, x-1, x-2 ....1 까지의 곱

# 정수 입력 : 3
# 결과 : 6
def fact_func(num) :   
    result = 1;
    while num >= 1 :
        result *= num
        num -= 1
        
    return result
#############################################

while True :
    num = int(input("정수 입력(0은 종료)"))
    
    if num == 0 :
        print("종료")
        break
    else : 
        result = fact_func(num)
    
    print("결과는 : %d"%result)
    print("*"*20)
```

```python
# 6(완료)
#자리수의 합
# 예를 들어 123 입력시 1+2+3을 계산하는 프로그램

def one_sum(num1) :
    arr = list()
    arr = list(num1)
    sum = 0
    for i in arr :
        sum += int(i)
    return sum
##########################################################

while True :
    num1 = str(input("정수 입력(0은 종료) : "))
    if num1 == '0' :
        print("*"*10)
        print("*  종료  *")
        print("*"*10)
        break
    
    sum = one_sum(num1)   
    print()
    print("결과는 {} 입니다.".format(sum))
    print()
```

```python
# 7
#숫자 맞추기 게임
# 1. 1부터 100 사이의 숫자를 맞추기
# 2. 숫자입력 : 30
# 3. 결과 : 업, 숫자입력 : 50
# 4. 결과 : 다운, 숫자입력 : 44
# 5. 정답. (시도횟수는 x번 입니다.)

def updown_func(num) :
    global cnt
    global red_cnt
    global ex_ans
    global msg
    global ans

    if msg == "업" and ex_ans > num :
        print("")
        print("")
        print("업이라니까? 생각이 있습니까 휴먼?")
        print("")
        print("괘씸죄로 시도 횟수 10회 추가됨")
        print("")
        print("")
        msg = "업"
        red_cnt += 10
        
    #다운인데 숫자를 크게 입력
    elif msg == "다운" and ex_ans < num :
        print("")
        print("")
        print("다운이라구요. 숫자부터 다시 익히는게 어떻습니까")
        print("괘씸죄로 시도 횟수 10회 추가됨")
        print("")
        print("")
        msg = "다운"
        red_cnt += 10
    
    elif num < ans :
        print("업")
        msg = "업"
        ex_ans = num
        cnt += 1 
    elif num > ans :
        print("다운")
        msg = "다운"
        ex_ans = num
        cnt += 1
    return


def ans_func():
    global cnt
    global red_cnt
    cnt += 1
    print("")
    print("정답입니다.")
    print("*"*20)
    print("시도 횟수는 : {}번 (벌점 {} 추가)".format((cnt+red_cnt), red_cnt))
    print("*"*20)
    return



######################################################

import random

num = 0        #입력값
cnt = 0        #시도 횟수
red_cnt = 0    #벌점
ex_ans = 0     #이전의 정답
msg = ""       #이전의 메세지
ans = random.randint(1,100)   #맞춰야할 숫자
    
while True :
    #맞출 숫자 입력
    num = int(input("숫자를 입력하세요(0은 종료) : "))
    
    #숫자가 맞을 시
    if num == ans :
        ans_func()
        break
        
    #종료    
    elif num == 0 :
        print("종료")
        break

    else :
        updown_func(num)
        
    
    
   # 업 다운 반대로 할 시 코멘트
```
