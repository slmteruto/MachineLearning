## 특징

    1) 묶음(block) - 변수, 함수를 묶어주는 단위
        - 여러 개의 함수를 묶어서 좀 더 큰 단위의 모듈로 관리
        - 변수 : 멤버변수, 속성
            ex)  ob1.a     a를 어트리뷰트라 함
        - 함수 : 멤버함수, 메소드          #속하지않은 함수를 함수라 하고 클래스 안의 함수를 메소드라고 하자
        
    2) 여러 개의 함수를 묶어서 좀 더 큰 단위의 모듈로 관리
    
    3) 설계도(추상화) -> 객체(구현), 인스턴스
    
    4) 사용자 정의 자료형
    
    5) 객체의 조건 : 
        - 상속성(Inheritance)   : 재사용성 극대화
        - 다형성(polymorphism)  :
        - 캡슐화(Encapsulation) :
        
    6) 객체 설계 5원치 : SOLID
        - SRP(단일 책임 원칙)       : 소프트웨어의 설계 부품(클래스, 함수 등)은 단 하나의 책임만을 가져야 한다
        
        - OCP(개방-폐쇄 원칙)       : 기존의 코드를 변경하지 않고(Closed) 기능을 수정하거나 
                                     추가할 수 있도록(Open) 설계해야 한다
                                     
        - LSP(리스코프 치환 원칙)    : 자식 클래스는 부모클래스에서 가능한 행위를 수행할 수 있어야 한다
        
        - DIP(의존 역전 원칙)       : 의존 관계를 맺을 때, 변화하기 쉬운것 보단 변화하기 어려운 것에 
                                     의존해야 한다는 원칙이다
                                     
        - ISP(인터페이스 분리 원칙)  : 한 클래스는 자신이 사용하지 않는 인터페이스는 구현하지 말아야 한다. 하나의 
                                     일반적인 인터페이스보다는, 여러 개의 구체적인 인터페이스가 낫다

## 클래스 작성 문법
    
    1. self
        - 현재 사용중인 객체의 주소를 저장하고 있는 참조 변수
        - read only
        
    2. 생성자와 소멸자 (소멸자는 거의 안씀)
        우리가 클래스를 만들고 계속 선언해버리면 메모리가 감당하기 힘들어지므로 
        
        - 메소드이다.
        - 객체를 생성하는 용도   
        - 생성자나 소멸자 둘다 객체를 처음만들때 단 한번만 호출
        - 생성자는 그래서 초기화 용도로 많이 씀.
        - 소멸자는 마무리작업으로 씀.
        
    3. static(객체를 뜻하는게 아님)
        - 동일한 공간을 모든 객체들이 메모리를 공동으로 사용 
        
    4. _, __ 
        1) _ : 기본 인터프리터(명령프롬프트)에서 사용, 최근사용한 값을 _ 라는 이름으로 저장된다.(연산도 가능)
             : 값을 무시할 때(자주사용)
             : 변수명과 함께 사용(접근 권한), __와 함께 사용
             
        2) __ :
    
    5. 상속
        - 상속 설계시 고려할 점
            1) 부모 클래스는 최소한의 코드로 작성
            2) 중복된 코드가 있어서는 안됨 (부모 자식간)
            3) 상속받은 자식 클래스는 반드시 기능을 확장(추가)시켜야 한다.
        
        - Override(재정의) 고쳐서 쓸 때, Overload(중복정의) 파이썬에선 지원되지 않는다.
        
        - super() : 부모객체에 접근하기 위한 주소를 리턴
        
        - Overriding을 이용한 다현성(표준화) 구현
        
        - 다중 상속
        
        - 추상클래스(Abstract Class)
            : 기본적으로 추상메소드를 한 개 이상 가지고 있어야 한다.
              다른 클래스에 추상메소드 하나만 껴 있어도 추상클래스 된다는 소리.
            : 추상메소드란 내용이 없는 메소드
            : 강제적으로 override 하게 해야됨
            : 추상클래스가 되면 자동으로 객체를 생성할 수가 없게 된다.
            : 일을 할 수 없게 된다.
            
        - object : 파이썬 모든 클래스의 부모클래스, 앞으로 하나의 클래스만 보지말고 상위 클래스도 확인
        object가 무슨기능을 하는지 찾아봐라

```python
# 예제 전자계산기 네번째 Class로 설계하기


## 계산기 기능 클래스
class Cal_clss :
    
    #체크 받고 넘어온 입력 숫자 두개 저장
    def __init__(self, inum1, inum2) :
        self.onum1 = inum1
        self.onum2 = inum2
    
    #저장된 숫자 두개 연산
    def cal_func(self, func) :
        if func == "+" :
            result = self.onum1 + self.onum2
            print()
            print("{} {} {} = {} 입니다.".format(self.onum1, func, self.onum2, result))
            print("-" * 30)
        elif func == "-" :
            result = self.onum1 - self.onum2
            print()
            print("{} {} {} = {} 입니다.".format(self.onum1, func, self.onum2, result))
            print("-" * 30)
        elif func == "*" :
            result = self.onum1 * self.onum2
            print()
            print("{} {} {} = {} 입니다.".format(self.onum1, func, self.onum2, result))
            print("-" * 30)
        elif func == "/" :
            result = self.onum1 / self.onum2
            print()
            print("{} {} {} = {} 입니다.".format(self.onum1, func, self.onum2, result))
            print("-" * 30)
        return 

    
## 숫자 체크 클래스
class Num_check :

    # 숫자인지 아닌지 판별하는 함수
    def isNumber(self, s):
        try:
            float(s)
            return True
        except ValueError:
            return False
    
    # 문자, 잘못된 값 구분하는 함수
    def numCheck(self, func) :
        while True :
            num1 = input("첫번째 숫자 입력")
            
            #숫자인지 아닌지 판단
            if self.isNumber(num1) :  
                if func == "/" :
                    while True :
                        num2 = input("두번째 숫자 입력(0불가)")
                        if num2 == "0" :
                            print("-"*50)
                            print("경고!! 0이 아닌 정수만 입력해주세요")
                            print("-"*50)
                        elif self.isNumber(num2) :
                            break
                        else :
                            print("-"*50)
                            print("경고!! 0이 아닌 정수만 입력해주세요")
                            print("-"*50)
                    break
                        
                else :
                    while True :
                        num2 = input("두번째 숫자 입력")
                        if self.isNumber(num2) :
                            break
                        else :
                            print("-"*50)
                            print("경고!  정수만 입력해주세요")
                            print("-"*50)
                    break

            else :
                print("-"*50)
                print("경고!  정수만 입력해주세요")
                print("-"*50)
            
        return int(num1), int(num2)



    
    
    
## 실행

while True :
    
    #클래스 선언
    chk = Num_check()
    
    
    #연산자 리스트
    func_list = ["+", "-", "*", "/"]
    
    #연산자 입력
    func = input("연산자를 입력해주세요(+,-,*,/)(x,q는 종료)")
    
    #연산자 컨트롤
    if func == "x" or func == "q":
        print()
        print("*"*15)
        print("종료")
        print("*"*15)
        print()
        break
        
    elif func in func_list :
        num1, num2 = chk.numCheck(func)  ##숫자체크
        cal = Cal_clss(num1, num2)       ##계산기 클래스 선언 (생성자값 위에서 체크한 num1,num2 부여)
        cal.cal_func(func)               ##연산 함수 실행
        
    else :
        print("-"*30)
        print("잘못입력하셨습니다.")
        print("-"*30)
        
    
```

```python
#static  (위의 person 클래스 재사용)

class Person :
    nation = "한국"
    name = str()
    age = int()
    height = float()
    telephone = str()
    
    #얘는 객체를 생성했을때만 메모리에 올라간다.
    def display(self) :
        print("국적 : ", self.nation)
        print("이름 : ", self.name)
        print("나이 : ", self.age)
        print("신장 : ", self.height)
        print("\n-----------------------")
        
    def test() :
        print("스태틱 테스트용")
    
    
    #파이썬에서 제공하는 고급함수 데코레이터, 확실하게 static 메소드라는걸 명시해준다.
    #저장시키기 위한 함수
    @staticmethod        
    def tel(num) : #유일하게 self를 쓰지 않는 경우
        #print("대표 전화 : {}({})".format(num, Person.nation))
        Person.telephone  = num    #self를 쓸수가 없다. 
        
    # 이경우 이 클래스의 멤버라고 하진 않는다. 스태틱은 객체와 상관없이 만들어지는것
    # self를 안쓰고 만들면 프로그램이 실행될때 이미 메모리에 올라가게 된다.
    # 객체를 쓰지않고도 사용할 수 있다.
    
    
    #저장된 것을 보여주는 함수
    @staticmethod
    def tel_show() :
        print("대표 전화 : {}({})".format(Person.telephone, Person.nation))
```

```python
#데코레이터 안쓰는 방법

class Counter :
    m_cnt = 0  #객체가 몇개 생성되었는지 카운트하는 변수
    
    
    def __init__(self) :
        Counter.m_cnt += 1
        

    def printCount() :
        print("객체의 수 : ", Counter.m_cnt)
    
    def printCount2(c) :
        print("객체의 수 : ", c.m_cnt)
        
    # 데코레이터를 안쓰고 쓰는 방법  
    # 가독성이 훨씬 떨어지므로 데코레이터를 사용한다.
    # 함수 이름을 내 마음대로 바꿀 수 있다는 장점은 있다.
    s1 = staticmethod(printCount)  # printCount를 staticmethod로 등록하겠다.
    
    s2 = classmethod(printCount2)  # printCount2 를 classmethod로 등록하겠다.
```

```python
#전자계산기 상속기능을 이용

class Calculator : 
    def __init__(self, first=None, second=None) :
        self.first = first
        self.second = second
        
    def setData(self, first=None, second=None) :
        self.first = first
        self.second = second
    
    def sum(self) :
        return self.first + self.second
    
    def sub(self) :
        return self.first - self.second
    
    def mul(self) :
        return self.first * self.second
    
    def div(self) :
        return self.first / self.second
    
class CalcV1(Calculator) : 
    def pow(self) :
        return self.first ** self.second
    
    # 상속받았기 때문에 first를 사용할 수 있음,
    # 그런데 부모가 이미 div를 줬는데 div2를 만들면 낭비이다.
    # 그래서 div 똑같은 이름으로 만들면 재정의로써 이게 override이다.
    def div(self) :
        if self.second == 0 :
            return "0으로 나눌수 없다."
        else :
            return self.first / self.second
    
    
    
```

```python
calc1 = Calculator()

calc1.setData(10,20)

print(calc1.sum())

calc2 = Calculator(20,30)

print(calc2.mul())

calc3 = CalcV1()
calc3.setData(30,4)
print(calc3.div())

calc4 = CalcV1(2,10)  ##부모의 없던 기능을 사용할 수 있다.
print(calc4.pow())

calc5 = CalcV1(40,0)
print(calc5.div())
```

```python
# __의 활용 예, 상속관련

class Person :
    say = "난 사람이야"
    __age = "20"  # 이렇게 해놓으면 부모클래스 외에서는 바꿀 수 없게 된다.
    #막아놓는다고 장땡이 아니라 사용할 수 있게 해줘야 하는데 이때 메소드를 쓴다.
    #이런 변수를 본다면 가급적 변경하거나 직접 사용하는것은 자제하도록 하자.
    
    def __init__(self, age) :
        print("Person 클래스")
        self.__age = age
        
    def info(self) :
        print("나이 : {}, 하고싶은말 : {}".format(self.__age, self.say))
    
    #접근하게 하는 메소드  getter 메소드라고도 한다.
    def getAge(self) :
        return self.__age

    #수정할수 있게 하는 메소드  (검증까지 포함시켜 주는게 좋다.)  setter 메소드라고도 한다.
    def setAge(self, age) :
        #검증
        self.__age = age
        
 #####################################################
class Employee(Person) :
    say = "일하는 동물"
    subject = "근로자"
    
    def __init__(self) :
        print("Employee 클래스")
        
    def einfo(self) :
        #print(self.__age)  ## 자식이라 할지라도 물려주지 않는다는것을 확인할 수 있다.
        print(self.getAge())  ## 이렇게 메소드를 이용해야 불러올 수 있음.
        print(self.say)
```

```python
# 예제 
# 상속, classmethod, staticmethod

# staticmethod일 경우 상속을 해도 부모의 리소스를 먼저 접근(오버라이딩 불가)
# classmethod일 경우에는 오버라이딩이 가능


#부모 클래스
class InTest : 
    t = "내가 상속해 주었음"
    
    
#     @staitcmethod  # 만약 아무것도 안적었다면 그냥 메소드이다. 그리고 (self를 적었어야 한다.)
#     def display() :
#         return InTest.t      ##일반 메소드였으면 self.t가 됐을 것이다.
    
    @classmethod # 만약 클래스메소드를 사용한다면
    def display(cls) :       ##이렇게 변수를 사용하면 Class이름이 변수로 넘어가진다.
        return cls.t
    
    


#자식 클래스
class ChildTest(InTest) :
    #상속받았기 때문에 변수t, 함수 display가 있다. 
    
    #override
    t = "내가 상속 받았음"
    

################################

#부모 출력
print(InTest.display())    ## 이렇게 사용하면 InTest가 display()의 변수로 넘어가게 되는게 classmethod

#자식 출력
print(ChildTest.display())
    
    
```

```python
#Overriding으로 뭘 할 수 있나

# 다형성(표준화)을 구현
# 다형성이란 여러가지 기능을 하나의 이름으로 사용할 수 있게 함.

# 예로 스마트폰을 만든다고 하자 

#부모 클래스 (최상위 느낌으로, 기본기능만 우선 구현)
class Phone : 

    def call(self) :
        print("전화 걸기")
        

#자식 클래스1
class FolderPhone(Phone) : 
    
    #고치기 위해 (overriding 하기 위해)
    def call(self) :
        print("통화 기능")
        
#자식 클래스2
class CameraPhone(Phone) :
    
    def call(self) :
        print("통화 기능 + 카메라 추가")
        
#자식 클래스3
class SmartPhone(Phone) :
    
    def call(self) :
        print("통화 기능 + 카메라 추가 + 인터넷 기능")     
        
        
p = Phone()
p.call()

p = FolderPhone()
p.call()

p = CameraPhone()
p.call()

p = SmartPhone()
p.call()
```

```python
## 예제 다중상속 

#1. 부모클래스에 Power 클래스 추가해서 다중상속을 구현해 본다.
#2. 전원을 켰을때만 볼륨 조절을 할 수 있게 한다.
#3. 볼륨 조정은 max값을 100, min값을 0으로 한다.

#결과 예시 :
#     1. TV
#     2. RADIO
#     3. SPEAKER
#     4. 종료
#     메뉴 : 1
        
#           -----TV-----
#         "현재 상태는 on"
#         1. 전원 on
#         2. 볼륨조정
#         3. 나가기

#            ----볼륨조정----
#            볼륨값 : 10
#              현재 TV의 볼륨은 10입니다.



#부모클래스1
class Volume:
        
    vol = 0
    def volumeUp(self, vol):
        pass
            
    def volumeDown(self, vol):
        pass
            
#부모클래스2
class Power :    
    def powerOn(self) :
        print("전원이 켜졌습니다.")
        return True
    
    def powerOff(self) :
        print("전원이 꺼졌습니다.")
        return False
        
        
############################################

class TV(Volume, Power):
    power_status = False
    power_message = "Off"    
    
    def powerOn(self) :
        print("TV 전원 On")
        return True, "On"

    def powerOff(self) :
        print("TV 전원 Off")
        self.vol = 0               #전원을 끄면 볼륨 0으로 초기화
        return False, "Off"
    
    
    def volumeUp(self, vol):
        self.vol += vol
        print("-"*70)
        print("TV 볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d 입니다."% (vol, self.vol))
        print("-"*70)


            
    def volumeDown(self, vol):
        self.vol -= vol
        print("-"*70)
        print("TV 볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d 입니다."% (vol, self.vol))
        print("-"*70)

            
            
############################################

class Radio(Volume, Power):
    power_status = False
    power_message = "Off"  
    
    def powerOn(self) :
        print("Radio 전원 On")
        return True, "On"

    def powerOff(self) :
        print("Radio 전원 Off")
        self.vol = 0
        return False, "Off"
    
    def volumeUp(self, vol):
        self.vol += vol
        print("-"*70)
        print("Radio 볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d 입니다."% (vol, self.vol))
        print("-"*70)

        
    def volumeDown(self, vol):
        self.vol -= vol
        print("-"*70)
        print("Radio 볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d 입니다."% (vol, self.vol))
        print("-"*70)

        
        
###########################################

class Speaker(Volume, Power):
    power_status = False
    power_message = "Off"  
    
    def powerOn(self) :
        print("Speaker 전원 On")
        return True, "On"

    def powerOff(self) :
        print("Speaker 전원 Off")
        self.vol = 0
        return False, "Off"
    
    def volumeUp(self, vol):
        self.vol += vol
        print("-"*70)
        print("Speaker 볼륨의 크기는 %d만큼 증가하여 현재 볼륨은 %d 입니다."% (vol, self.vol))
        print("-"*70)

        
    def volumeDown(self, vol):
        self.vol -= vol
        print("-"*70)
        print("Speaker 볼륨의 크기는 %d만큼 감소하여 현재 볼륨은 %d 입니다."% (vol, self.vol))
        print("-"*70)


#############################################
class IT_class : 
    menu_list = ["1", "2", "3", "4"]

    tv = TV()
    radio = Radio()
    speaker = Speaker()    

    #메뉴 선택시 그 메뉴 객체를 담기 위함 (tv, radio, speaker)
    select = ""
    
    ##시작 메뉴 선택
    def select_it(self) :
        
        while True :
            print("---메뉴선택---")
            print("1. TV")
            print("2. Radio")
            print("3. Speaker")
            print("4. 종료")
            print("-------------")
            menu = input("사용할 메뉴")
            print()
            
            #메뉴에 없는것 선택 시
            if menu not in self.menu_list :
                print("*"*50)
                print("경고! 잘못입력하셨습니다.")
                print("*"*50)
                print()

            #메뉴 선택
            elif menu in self.menu_list :
                if menu == "1" :
                    print("TV 선택")
                    print()
                    select = self.tv
                    self.control(select)
                elif menu == "2" :
                    print("Radio 선택")
                    print()
                    select = self.radio
                    self.control(select)
                elif menu == "3" :
                    print("Speaker 선택")
                    print()
                    select  = self.speaker
                    self.control(select)
                elif menu == "4" :
                    print("종료")
                    print()
                    break
            print()

            
    # 세부 메뉴 선택 함수
    # select는 객체주소
    def control(self, select) :  
        while True :
            print("현재상태는 : ", select.power_message)
            print("무엇을 하시겠습니까")
            print("-"*20)
            print("1. 전원 켜기/끄기")
            print("2. 볼륨 설정")
            print("3. 나가기")
            print()
            menu = input("선택 : ")

            if menu == "1" :
                print()
                #전원이 켜져있다면 끄기버튼으로 작동
                if select.power_status :
                    select.power_status, select.power_message = select.powerOff()
                #전원이 꺼져있다면 켜기버튼으로 작동
                else :
                    select.power_status, select.power_message = select.powerOn()
                print()
                
            elif menu == "2" :
                print()
                print("---볼륨 설정---")
                print("현재 볼륨 : ", select.vol)
                print()
                
                #전원이 켜져있을때만 활성화
                if select.power_status :
                    print("1. 볼륨 증가")
                    print("2. 볼륨 감소")
                    print("3. 뒤로")
                    print()
                    vol_menu = input("선택 : ")
                    if vol_menu == "1" :
                        vol_val = int(input("몇으로 하시겠습니까"))
                        if 0 <= (select.vol + vol_val) <= 100 :
                            select.volumeUp(vol_val)
                        else : 
                            print("*"*50)
                            print("경고! 범위를 초과했습니다.")
                            print("*"*50)
                        print()
                    elif vol_menu == "2" :
                        vol_val = int(input("몇으로 하시겠습니까"))
                        if 0 <= (select.vol - vol_val) <= 100 :
                            select.volumeDown(vol_val)
                        else :
                            print("*"*50)
                            print("경고! 범위를 초과했습니다.")
                            print("*"*50)
                        print()
                    elif vol_menu == "3" :
                        break
                    else :
                        print("*"*50)
                        print("경고! 잘못 선택하셨습니다.")
                        print("*"*50)
                        print()
                else :
                    print("*"*50)
                    print("경고! 전원이 꺼져있습니다.")
                    print("*"*50)
                    print()
            elif menu == "3" :
                print()
                break
            else :
                print("*"*50)
                print("경고! 잘못 선택하셨습니다.")
                print("*"*50)
                print()

################################################

# 메뉴 실행기
IT_class().select_it()
```

```python
# 추상클래스, 추상메소드

import abc                ##abc모듈 (ABstract Class라는 뜻)

#부모 클래스
class Volume2(metaclass=abc.ABCMeta) :  ##추상메소드로 만들었다면 클래스도 추상클래스로 해주어야 한다.
    vol = 0
    
    #abstract로 만들기 위한 데코레이터
    @abc.abstractmethod
    def volumeUp(self, vol) :
        pass
    
    @abc.abstractmethod
    def volumeDown(self, vol) :
        pass
    

#자식 클래스
class TestVolume2(Volume) :
    
    def volumeUp(self, vol) :
        self.vol += vol
        print("소리증가 : ", self.vol)

        #override 안해도 에러는 안난다. abstract가 아닌 일반 메소드이기 때문에
        #그렇지만 부모클래스에서 추상메소드로 만들어버리면 에러가 난다.
    def volumeDown(self, vol) :
        self.vol -= vol
        print("소리감소 : ", self.vol)
        
#######################################
#t3 = Volume2()    #추상클래스가 되면 객체 생성이 불가능하다.
#t3.volumeUp(100)

t4 = TestVolume2()
t4.volumeUp(100)
```

```python
# 객체별로 변수 추가 금지 : __slot__

class Animal :
#     name = "동물"
#     age = 0
    __slots__ = ["name", "age"]  #이렇게하면 이 클래스에서는 반드시 이 두개의 변수만 사용해야된다.
    
    def display(self) :
        print(self.name, self.age)
        
        
###################################
# 객체별로 따로 값을 보관하는 방법

a = Animal()
b = Animal()

#위에서 __slot__을 써주면 반드시 초기화를 시켜줘야 한다.
a.name = "멍멍이"
a.age = 10
b.name = "야옹이"
b.age = 20

a.display()
b.display()


#a.etc = "예방접종 받았음"  # 변수를 임시로 추가하는 방법

#print(a.etc)
#print(b.etc)  #b에서는 당연히 쓸 수 없다.
```

## 클래스의 관계

    1) is a 관계 (상속관계, 일반화관계)
        특정 클래스 객체가 다른 클래스 객체에게 자신의 멤버를 물려주는 상속 관계
        강한 결합관계
        UML : 비어있는 화살표
        
    2) has a 관계(연관관계, 포함관계)
        특정 클래스 내에서 해당 클래스의 멤버처럼 다른 클래스 객체를 포함하고 있는 경우
        - Aggregation : 약한 관계
        
        - Composition : 강한 관계 (그 클래스가 있어야만 쓸 수 있는 정도) 
        
        UML : 다이아모양 화살표 (강한건 검은색, 약한건 비어있는)
        
    3) UML 

```python
# 약한관계 강한관계 판단

# Handle과 Car가 상속관계도 아닌데 무슨 관계가 있냐에 주의

class Handle : 
    quantity = 0  #핸들 각도, 몇도냐에 따라 좌,우 구분
    
    def leftTurn(self, q) :
        self.quantity = q
        return "좌회전"
    
    def rightTurn(self, q) :
        self.quantity = q
        return "우회전"
    
    def straight(self, q) :
        self.quantity = q
        return "직진"
    
class Car :
    turnShow = "정지"
    
    def __init__(self, ownerName) :
        self.ownerName = ownerName
        self.handle = Handle()  #Handle과 Car의 관계
        
    def turnHandle(self, q) :
        if q > 0 :
            self.turnShow = self.handle.rightTurn(q)
        elif q < 0 :
            self.turnShow = self.handle.leftTurn(q)
        elif q == 0 :
            self.turnShow = self.handle.straight(q)
            
            
#main

tom = Car("tom")
tom.turnHandle(10)
print(tom.ownerName + "의 회전량은 " + tom.turnShow + str(tom.handle.quantity))
```

```python
# 냉장고   

# 강한관계, 약한관계 
# 보기에 따라 다를 수 있다.
# 객체를 어떻게 포함했냐라는 패턴에 주의

class Fridge :
    isOpen = False
    foods = []
    
    def open(self) :
        self.isOpen = True
        print("냉장고 문열기")
    
    def close(self) :
        self.isOpen = False
        print("냉장고 문닫기")
    
    def put(self, thing) :
        if self.isOpen :
            self.foods.append(thing)
            print("냉장고에 음식이 들어감")
            self.list()
        else :
            print("냉장고 문이 닫힌 상태")
    
    def list(self) :
        for f in self.foods :
            print("-", f.irum, f.ex_data)
        print()
    
    

class FoodData :
    def __init__(self, irum, ex_data) :
        self.irum = irum
        self.ex_data = ex_data
        
############################################

f = Fridge()
apple = FoodData("사과", "2019-12-16")

f.open()
f.put(apple)
f.list
f.close()
    
```
