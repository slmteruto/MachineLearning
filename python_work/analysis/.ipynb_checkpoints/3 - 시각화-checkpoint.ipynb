### 산포도 (scatter plot)

    두 변수의 관계를 확인
        - 양, 음
            양 : 하나의 데이터가 커질때 다른 데이터도 커짐  (비례)
            음 : 하나의 데이터가 커질때 다른 데이터는 작아짐(반비례)
            
            
    산포도만으로 확인할 수 없는 데이터가 있기 때문에 box graph를 사용한다.

```python
x1 = np.random.normal(1, 1, size = (100, 1))
x2 = np.random.normal(-2, 4, size = (100, 1))

x = np.concatenate((x1, x2), axis = 1) 
print(x[:5])

df3 = pd.DataFrame(x, columns = ["x1", "x2"])
df3.head()

# df3.plot.scatter("x1", "x2")
plt.scatter(df3["x1"], df3["x2"])
```

### 상자그래프(box plot)

```python
df3.plot.box()

#그래프에 선이 그어져있는것은 중앙값이다. 
#막대선은 최대 최소값
```

```python
# 절차를 지켜서 그래프를 그리기

tips = sns.load_dataset("tips")
tips.head()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)  # 1행 1열 1개의 그래프 (축의 정보)  1,1,1이 기본값

#total_bill에 대한 금액 분포를 확인
ax.hist(tips["total_bill"], bins=20)    # 이렇게만 하면 보는사람이 헷갈린다. x,y축을 모르니

ax.set_title("식사 대금에 대한 히스토그램")  #그래프 이름
ax.set_xlabel("frequency")                 # x축 이름
ax.set_ylabel("total Bill")                # y축 이름
```

```python
# Tip과 Total_bill과의 관계를 알고 싶다 


fig = plt.figure()
ax = fig.add_subplot(1,1,1)  

ax.scatter(tips["total_bill"], tips["tip"])    # 두 변수를 보낸다. 

ax.set_title("식사 대금과 팁에 대한 산포도")  #그래프 이름
ax.set_xlabel("total_bill")                 # x축 이름
ax.set_ylabel("tip")                # y축 이름

#방향이 있는 양의 관계이다/ 식사금액이 커질 수록 팁도 커진다. 
```

```python
# 성별, Tip의 관계

fig = plt.figure()
ax = fig.add_subplot(1,1,1)  

ax.boxplot([tips[tips["sex"]=="Female"]["tip"], tips[tips["sex"]=="Male"]["tip"]],
          labels=["Female", "Male"]) 
#list로 하나로 묶어서 보내야 함.

ax.set_title("성별에 따른 팁에 대한 boxplot")  #그래프 이름
ax.set_xlabel("")                 # x축 이름
ax.set_ylabel("")       

```

```python
#성별을 0,1로 표현
def record_sex(sex) :
    if sex == "Female" :
        return 0
    else :
        return 1
    
tips["sex_color"] = tips["sex"].apply(record_sex)
tips.head()

#matplot으로 표현
fig = plt.figure()
ax = fig.add_subplot(1,1,1)  

ax.scatter(tips["total_bill"], tips["tip"], c=tips["sex_color"], s=tips["size"]*10, alpha = 0.5)    
#c : 색상집어넣는 옵션
#s : 크기를 나타내기
#alpha : 투명도

ax.set_title("식사 대금과 팁에 대한 산포도")  #그래프 이름
ax.set_xlabel("total_bill")                 # x축 이름
ax.set_ylabel("tip")     

#####################################################################
#seaborn으로 표현
sns.lmplot(x="total_bill", y="tip", data=tips, fit_reg=False, hue="sex", 
          scatter_kws={"s":tips["size"]*10}, markers=["o", "x"]) 

    # hue 어떤 데이터를 표기할 건지 
    # scatter 것을 쓸 수 있다.  
    # markers 점 모양을 바꿀 수 있다.
```

```python
## 한글 분석
# http://konlpy.org   #한글 자연어 처리

# 설치 
#1. https://www.lfd.uci.edu/~gohlke/pythonlibs/#jpype  가서 최신 버전 cp37m-win_amd64 로 다운
#2. cmd 로  다운받은 폴더로 가서 pip install jpy어쩌구 저쩌구 
#3. !pip install konlpy

#사용
# from konlpy.tag import Twitter   #옛날이름
#또는 
from konlpy.tag import Okt  #최신이름 
from konlpy.corpus import kolaw


# 빈도수 계산해주는 모듈
import nltk  
```

```python
ko_text = kolaw.open("constitution.txt").read()
ko_text

#금지어 설정
stop_words = ["제" , "월", "일", "조", "수", "때", "그", "이", "바", "및", "안", "정", "위", "관"]

t = Okt()
ko_token = t.nouns(ko_text)  #ko_text중 명사만 뽑아줌
ko_token = [each_word for each_word in ko_token if each_word not in stop_words]  # 금지어 제외
print(len(ko_token))

import nltk  # 빈도수 계산해주는 모듈

ko = nltk.Text(ko_token)
```
