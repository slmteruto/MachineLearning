## 1) 데이터 수집


```python
# cctv자료 불러오기
cctv_seoul = pd.read_csv("data/seoul/cctv.csv")
cctv_seoul.head()
```

```python
# 인구수 자료 불러오기
pop_seoul = pd.read_excel("data/seoul/Report.xls")
pop_seoul.head()
```

```python
## 위 결과의 인덱스들이 맘에 안들어서 빼고 불러오기
pop_seoul = pd.read_excel("data/seoul/Report.xls", header=2)  #header : 두번째를 제목으로 선정, 첫번째는 빠짐
pop_seoul.head()
```

```python
## 엑셀은 컬럼이 A, B, C... 로 나뉘어 있다. 원하는 컬럼 가져올때
#usecols : 내가 가져올 컬럼만 선택가능(엑셀만)
pop_seoul = pd.read_excel("data/seoul/Report.xls", header=2, usecols = "B, D, G, J, N") 
pop_seoul.head()
```

## 2) 전처리

```python
# 컬럼 이름을 변경

# cctv자료에서 기관명이라는 컬럼명을 "구별"로 변경
cctv_seoul.rename(columns = {cctv_seoul.columns[0] : "구별"}, inplace = True) #inplace는 실제 데이터에 바로적용
cctv_seoul.head()
```

```python
# 인구수 데이터는 컬럼명을 구별, 인구수, 한국인, 외국인, 고령자 로 수정
pop_seoul.rename(columns = {pop_seoul.columns[0]:"구별", pop_seoul.columns[1]:"인구수",
                           pop_seoul.columns[2]:"한국인", pop_seoul.columns[3]:"외국인"
                           ,pop_seoul.columns[4]:"고령자"}, inplace=True)
pop_seoul.head()
```

## 3) 탐색적 데이터 분석

```python
# 탐색적 데이터 분석은 데이터를 파악하는 것

# 어느 구가 가장 CCTV가 많이 설치 되었는가
#-------------------------------------------

#CCTV 데이터에서 소계를 기준으로 정렬 
cctv_seoul.sort_values(by = "소계", ascending=False).head(5)
cctv_seoul.sort_values(by = "소계", ascending=True).head(5)
```

```python
# CCTV의 설치 증가율
#-------------------

#CCTV 설치 연도별 증가율을 구해서  "최근 증가율" 컬럼 추가
# ((2014+2015+2016) / 2013년도 이전) * 100
cctv_seoul["최근 증가율"] = ((cctv_seoul["2014년"] +cctv_seoul["2015년"]+cctv_seoul["2016년"])/cctv_seoul["2013년도 이전"]) *100

cctv_seoul.sort_values(by = "최근 증가율", ascending=False).head()  
cctv_seoul.sort_values(by = "최근 증가율", ascending=True).head()

```

```python
## 인구대비 CCTV 설치 현황을 알아보자 
#----------------------------------

#인구수 데이터에 필요없는 행 제거 (서울시 전체 합계)한다.      del, drop
pop_seoul.drop([0], inplace=True)  #drop은 행기준이다. 
pop_seoul.head()
```

```python
# 구별 데이터의 중복이나 결측치등을 확인  (unique)
pop_seoul["구별"].unique()
pop_seoul[pop_seoul["구별"].isnull()]     #null값 데이터 확인  이 행도 삭제

```

```python
pop_seoul.drop([26], inplace=True)
```

```python
pop_seoul["구별"].unique()
```

```python
# 어느 구가 가장 인구수가 많은가
pop_seoul.sort_values(by = "인구수", ascending=False).head()

pop_seoul.sort_values(by = "한국인", ascending=False).head()

pop_seoul.sort_values(by = "외국인", ascending=False).head()

pop_seoul.sort_values(by = "고령자", ascending=False).head()


```

```python
# 외국인 비율, 고령자 비율 컬럼 추가 


#외국인비율
pop_seoul["외국인 비율"] = pop_seoul["외국인"] / pop_seoul["인구수"] * 100
pop_seoul["고령자 비율"] = pop_seoul["고령자"] / pop_seoul["인구수"] * 100

pop_seoul.sort_values(by="외국인 비율", ascending =True).head()

pop_seoul.sort_values(by="고령자 비율", ascending =True).head()
```

```python
#인구 대비 CCTV 설치 현황을 알아보기
#---------------------------------------

#cctv데이터와 인구수 데이터 합치고 분석
data_result = pd.merge(cctv_seoul, pop_seoul, on = "구별")
data_result.head()


#필요없는 컬럼 정리(제거) : 2013년도 이전, 2015, 2016 제거           del, drop 행기준(drop), 컬럼기준(del)
#여기서는 컬럼이니까
del data_result["2013년도 이전"]
del data_result["2014년"]
del data_result["2015년"]
del data_result["2016년"]
data_result.head()
```

```python
# 위 데이터에서 구별컬럼을 인덱스로 설정

data_result.set_index("구별", inplace=True)
data_result.head()


```

```python
#CCTV 소계가 인구수랑 관련있는지, 외국인수랑 관련있는지, 고령자랑 관련있는지 확인

# 상관계수 : 두 변수의 관계를 수치로 나타냄   
# 그래프로 보고싶으면 산포도, 양의 방향, 음의 방향

# 상관계수 지정 :  -1  ~ 1     np.corrcoef()
# -1에 가까울수록 음의 상관관계
# 1에 가까울수록 양의 상관관계
# 0에 가까울수록 상관없음


print(np.corrcoef(data_result["고령자 비율"], data_result["소계"]))

print(np.corrcoef(data_result["외국인 비율"], data_result["소계"]))  #외국인 비율은 정말 상관이 없다.

print(np.corrcoef(data_result["인구수"], data_result["소계"])) 

#셋다 상관이 없었다. 
```

```python
data_result.sort_values(by="소계", ascending=False).head()
data_result.sort_values(by="인구수", ascending=False).head()
data_result.sort_values(by="인구수", ascending=True).head()
```

## 시각화

```python
#cctv가 많이 설치된 구별로 그래프 그리기 

data_result["소계"].plot(kind="bar")
#또는
data_result["소계"].plot.bar()
data_result["소계"].plot(kind="bar", figsize=(5,5), grid=True)  #좀더 크게 그리기, grid, 선긋기
```

```python
# data_result["소계"].plot(kind="barh", figsize=(5,10), grid=True) # 옆으로 눞히기 barh
data_result["소계"].sort_values().plot(kind="barh", figsize=(5,10), grid=True) #이렇게 하면 소계 기준으로
```

```python
# CCTV비율 컬럼 추가  (소계/인구수 * 100)
data_result["cctv비율"] = data_result["소계"]/data_result["인구수"] *100

data_result.head()

#위 토대로 그래프
data_result["cctv비율"].sort_values().plot(kind="barh", figsize=(5,10), grid=True)
```

```python
# cctv의 설치 대수와 인구수의 관계를 산포도로 확인 
plt.figure(figsize=(6, 6))
plt.scatter(data_result["인구수"], data_result["소계"], s= 50)  #결과를 보면 관계가 없는것을 알 수 있다.
plt.xlabel("인구수")
plt.xlabel("소계")
plt.grid()


#회귀분석은 분석에도 사용하지만 예측하는데도 찾는다. (인구수가 얼마정도 되면 cctv가 얼마나 늘어나겠다 이런것)
# y = wx + b  (w : 가중치) 
# 선을 찾는것 : 최소 제곱법
```

```python
# poly1d  : 1차방정식을 만들어주는 함수 

# 예시
a = np.poly1d([1, 2])
print(a)  #위에서 1, 2대신 변수를 넣어가지고 방정식을 만드는게 찾는 것(?)

b = np.poly1d([1,2,3])  # 2차 방정식
print(b)
```

```python
# 회귀선 작성(y = wx + b)
fp1 = np.polyfit(data_result["인구수"], data_result["소계"], 1)   #가장 적정한 w를 찾아줌  #숫자는 차원

print(fp1)   # 기울기(w), 절편(b)  으로 리턴

f1 = np.poly1d(fp1)  #fp1을 넘기면 방정식이 나온다
print(f1)

fx = np.linspace(100000, 700000, 100)  # 10~70만까지 100단위

plt.figure(figsize=(6, 6))
plt.scatter(data_result["인구수"], data_result["소계"], s= 50)  #결과를 보면 관계가 없는것을 알 수 있다.
# plt.plot(fx, 0.001104 x + 1069)  # 선그리기  x좌표, y좌표   #이렇게 하면 불편하니까
plt.plot(fx, f1(fx), lw=3, color="g", ls="dashed")  #iw: 선의 두께, color: 선의 색, ls : 라인스타일
plt.xlabel("인구수")
plt.xlabel("CCTV")
plt.grid()

#선에 가까워야 데이터가 제일 정확한 데이터/ 적정한 데이터 , 멀어질 수록 오차가 크다는 것
```

```python
# 각 데이터의 오차를 구해서 구 이름을 출력

data_result["오차"] = np.abs(data_result["소계"] - f1(data_result["인구수"])) #인구수에 대한 적정값
data_result.head()
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(data_result["인구수"], data_result["소계"], s= 50, c=data_result["오차"])  #오차에 따른 색바꿈
plt.plot(fx, f1(fx), lw=3, color="g", ls="dashed")  
plt.xlabel("인구수")
plt.xlabel("CCTV")
plt.grid()
plt.colorbar()

df_sort = data_result.sort_values(by="오차", ascending=False) #오차가 적은것에 관심있으니 
df_sort.head(10)

for n in range(10) :   #겹치는거 표현?
    plt.text(df_sort["인구수"][n], df_sort["소계"][n], df_sort.index[n], fontsize=10) 
    #text :그래프에 글자를 나타내는 함수
    
    
    
    
```
