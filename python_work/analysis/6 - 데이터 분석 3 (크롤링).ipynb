```python
"""
*.XML(Extensible Markup Language)

    1)메타언어
        GML -> SGML -> XML
    2)데이터를 위한 언어
    3)DB
    4)데이터의 표준화 : 이기종 시스템간의 정보교환, 웹서비스, 
        유비쿼터스, 사물인터넷 기타 등등...

*. XML 문법

    XML(.xml), DTD(.dtd), XML스키마(.xsd), XSL(XML Style Sheet Language, XSLT)

    (1) XML 문서 종류
        1) Well-Formed XML Document(잘 짜여진 문서)
                2) Valid XML Document(유효한 문서)

    (2) XML 기본 구조
        Prolog  - 1개
        Element - 1개
        Misc    - 0개 이상

    (3) Element 문법
        - 작성 규칙
            a) 모든 XML문서는 반드시 단 하나의 루트 엘리먼트를 가진다.
            b) 루트 엘리먼트는 여러 개의 자식 엘리먼트를 가질 수 있고 또 그 자식도 자신의 자식을 가질 수 있다.
            c) 시작태그와 끝태그는 반드시 짝을 이루어야 한다.
            d) "<"는 값으로 사용 불가. ">"는 사용할 수 있으나가급적 사용 금지
            e) "<"와 ">" 다음에 공백문자가 올 수 없으며 반드시 시작태그와 끝태그의 이름이 같아야 한다.
                    
        - 종류
            a) 내용을 가지는 엘리먼트
            b) 내용이 없는 엘리먼트

        - 내용에 대한 종류
            a) 문자 데이터
            b) 자식 엘리먼트
            c) 엔티티 또는 문자 참조(&엔티티명;)
            d) CDATA Section
            e) 프로세싱 지시자
            f) 주석
            g) 공백 문자열

    (4) DTD 
        - 종류
            내부 DTD
            외부 DTD
                
        - 구성 요소
            엘리먼트 선언
            속성(attribute) 선언
            Entity 선언
            Notation 선언
            프로세싱 지시자
            파라미터 엔티티 참조
            주석
            공백
            conditional section
                    
        - 문서 유형 선언
            <!DOCTYPE 루트엘리먼트명 SYSTEM 또는 PUBLIC "식별자">
            DOCTYPE은 반드시 대문자
            SYSTEM은 특정 단체나 업체내부에서 사용되는 경우
            PUBLIC은 공개
            식별자는 다운로드 받을 수 있는 경로
            PUBLIC일 경우 추가된 식별자
            +-//DTD를 개발 및 유지보수 업체명//DTD명 및 버전번호//사용된 언어

        - 엘리먼트 선언
            <!ELEMENT 엘리먼트명 컨텐트 유형>
            컨텐트 유형
                #PCDATA	: 내용으로 문자데이터만 갖는 엘리먼트
                자식 엘리먼트
                    , : 작성 순서
                        | : 선택
                        ? : 생략하거나 한번만 작성
                        + : 한번 이상
                        * : 생략하거나 여러번 작성
                        기호 없음 : 단 한번만 작성
                    EMPTY
                    MIXED : 문자데이터 또는 자식엘리먼트를 혼합형태
                    ANY

            - ATTRIBUTE 선언
                <!ATTLIST 엘리먼트명 속성명 속성유형 디폴트선언>
                속성 유형
                    CDATA     : 문자 데이터
                    ENUMERATION : dtd에 나열된 값 중 하나가 와야 함
                    ID : 유일한 값을 지정
                    IDREF/IDREFS : ID값을 참조
                    NMTOKEN/NMTOKENS : 이름 작성 규칙을 준수하는 
                        데이터만 사용
                    NOTATION : dtd에 명시적으로 선언된 notation만 사용가능
                    ENTITY : dtd에 명시적으로 선언된 entity만 사용 가능
        
        (5) 네임 스페이스
            CML
            -----
            <?xml version=1.0>
            <정보>
                <고유번호>111111-1111111</고유번호>
                <이름>홍길동</이름>
                ...
            </정보>
            
            PML
            -----
            <?xml version=1.0>
            <정보>
                <고유번호>LC100</고유번호>
                <이름>캠코더</이름>
                ...
            </정보>
            
            OML
            -----
            <?xml version=1.0>
            <주문정보 xmlns:고객="http://www.a.com/2017/Custom"
                xmlns:상품="http://www.a.com/2017/Product"
                xmlns="http://www.a.com/2017/Order">
                <주문번호>1</주문번호>
                <주문수량>10</주문수량>
                <결제>
                    <방법>현금</방법>
                    <금액>10000000</금액>
                </결제>
                
                <고객:고유번호>111111-1111111</고객:고유번호>
                <고객:이름>홍길동</고객:이름>
                
                <상품:고유번호>LC100</상품:고유번호>
                <상품:이름>캠코더</상품:이름>
                ...
            </주문정보>
"""
```

```python
import xml.etree.ElementTree as elemTree  #파이썬에서 XML을 다룰수있게 하는 모듈
```

```python
#XML을 파싱하는 방법

# 1. xml파일로 존재하는 경우 : parse()
# 2. xml문자열 형태로 존재하는 경우 : fromstring()
```

```python
#xml 불러오기 ElementTree, open

#open()  #이걸로도 열리긴 하는데 효율이 조금..

#메모리에 저장하고 가져다가 사용하는 방법
tree = elemTree.parse("data/users.xml")
tree

#메모리에 바로 생성해서 쓰는 경우
xmlstr = """<?xml version="1.0" encoding="utf-8" ?>
<users>
    <user grade="gold">
            <name>Kim Cheol Soo</name>
            <age>25</age>
            <birthday>19940215</birthday>
        </user>
        <user grade="diamond">
            <name>Kim Yoo Mee</name>
            <age>21</age>
            <birthday>19980417</birthday>
        </user>
</users>
"""
#<> 태그 하나하나가 요소(Element)가 된다.

tree = elemTree.fromstring(xmlstr)
tree
```

```python
#태그 검색및 데이터 다루기
#XPATH : ./  현재위치에서 찾겠다
#       ../ 상위위치에서 찾겠다
# user = tree.find("user")   #Tree 안에 여러개 있다면 배열로 저장이 된다. 
# user = tree.find("user[1]")  #두번째 user에 접근하고 싶다면
# 또는
user = tree.find("./user[1]")
user
dir(user) #목록 뿌리기 

print(user.tag) # 현재객체의 태그명
print(user.attrib) # dict형식으로 속성도 가져올 수 있다.
print(user.get("grade"))  #해당 값 

user_name = user.find("name")  #user안에서만 또 찾을 수 있는 방법
user_name
print(user_name.tag)
print(user_name.text)

user_age = user.find("age")
print(user_age.text)

```

```python
# 태그의 조건으로 검색

user = tree.find("user[2]")  #user태그에 접근하고 

user = tree.find("user[@grade]") #@로하면 속성에 바로 접근하겠다. 

user = tree.find("user[@grade][2]") #@로하면 속성에 바로 접근하고 두번째 인덱스로 찾아감

print(user.attrib)
```

```python
#여러개의 태그를 한꺼번에 가져오기 
users = tree.findall("./user")

for user in users :
    print(user.attrib)
    print(user.find("name").text)
```

```python
#어트리뷰트 값 읽기
tree = elemTree.parse("data/users.xml")
users = tree.getroot()
print(users)
print(users.tag)
print(users.attrib) #꼭대기에 있어서 아직 값이 나오진 않는다. 

user = users.find("./user")  #첫번째 유저태그접근
print(user.get("grade"))  #이젠 속성이 있어서 값이 나온다. 
print(user.keys())        # 키값을 가져오기
print(user.items())       # dict형식이라 dict함수 사용 가능하다. 
```

## JSON(JavaScript Object Notation)
    - JSON : XML이 적은 데이터에도 너무 많은 태그가 생기므로 태그를 줄인 데이터 위주형식

    - dumps : 데이터를 저장할 때
    - loads : 데이터를 불러올 때

```python
import json
```

```python
j1 = {"name":"홍길동", "birth":"0524", "age":"20"}
j2 = json.dumps(j1)
print(type(j2))
print(j2)
```

```python
json.dumps([1,2,3])
json.dumps([4,5,6])
```

```python
j3 = json.loads(j2)
print(type(j3))
print(j3)
```

## 실전예제

```python
#네이버 환율정보 가져오기
url = "https://finance.naver.com/marketindex/"
page = urlopen(url)
soup = BeautifulSoup(page, "lxml")

span = soup.find_all("span", class_="value")
print("미 환율 : ", span[0].get_text())


div = soup.find_all("div", class_="head_info")
span1 = div[0].find_all("span")
print(span1[0].string)

#css 접근
span2 = soup.select_one("div.head_info > span.value").string   
#css도 .접근(클래스접근)//  div의 head_info 접근
# > 그안의 span중 class = value인것 접근

print(span2)
```

```python
url = "https://ko.wikipedia.org/wiki/%EC%9C%A4%EB%8F%99%EC%A3%BC%22"
page = urlopen(url)
soup = BeautifulSoup(page, "html.parser")

div = soup.find_all("div", class_="mw-parser-output")
ul = div[0].find_all("ul")


ul1 = soup.select("div.mw-parser-output > ul")
ul1[0]
for i in ul1[0] :
    if i != "\n" :
        print(i.string)
```
