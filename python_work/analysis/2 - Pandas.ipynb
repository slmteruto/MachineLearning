## Series

```python
s = pd.Series([4684651, 6848, 86878, 246855])   # enumerate같이 인덱스화 시켜줌

print(s)


print(list(enumerate(s, start = 1)))  # 예전에 하던 방법
```

```python
# 인덱스를 자유롭게 지정
s = pd.Series([4684651, 6848, 86878, 246855], index = ["서울", "부산", "인천", "대구"])  
print(s)
print(type(s))
print("-------------")
print(s.values)
print(s.index)
print("-------------")
s.index.name = "도시"  # 인덱스에도 이름을 지정할 수 있다.
print(s)
```

```python
#백터 연산
s1 = s/1000000    #인덱스랑 값이 있는데 계산이 되나? 

print(s1)         # 값에만 연산이 된다
```

```python
# 인덱싱   데이터 분석에서 인덱스는 중요

print(s[1], s["부산"])
print("---------------")
print(s[[0, 3, 1]], s[["서울", "대구", "부산"]])
print("---------------")
print(s[(250e4<s) & (s<500e4)])   # 범위 지정 가능, 

print("---------------")
#슬라이싱

print(s[1:3])
print("---------------")
print(s.부산, ",", s.서울)
```

```python
# Series와 Dict

print("서울" in s)
print("대전" in s)

print("--------------------")
#그냥 출력
print(s.items())
for k, v in s.items() :
    print("%s = %d"%(k,v))

print("--------------------")
#dict형
s2 = pd.Series({"서울":9904312, "부산":5557757, "인천":3015684,"대전":1387651})
print(s2)
```

```python
# 인덱스 기반 연산
print(s)
print("--------------------")
print(s2)

print("--------------------")
ds = s2 - s
print(ds)                   # 인덱스 별로 계산
#Nan : Not a Number   (None과 같음)


print("--------------------")
print(s2.values - s1.values)  #인덱스 빼고 순서대로 값만 계산  그래서 위랑 값이 다름
```

```python
#결축치 제거
print(ds)
print("---------------------")
print(ds.notnull())      #True, False 구분하는 함수
print("---------------------")
print(ds[ds.notnull()])  #True인것만 출력

print("---------------------")

#인구 증가율 (( 끝연도-시작연도 / 시작연도) * 100)
ds2 = (s2-s)/s * 100
print(ds2)
```

```python
# 데이터를 수정, 삭제, 갱신

ds2["부산"] = 1.63
print(ds2)
print("-------------------------")
ds2["대구"] = 1.41
print(ds2)

#삭제
del ds2["서울"]
print(ds2)
```

## 고급 인덱싱 : indexer
    - loc : 라벨값 기반의 2차원 인덱싱을 지원하는 인덱서 (location)
    - iloc : 순서를 나타내는 정수기반의 2차원 인덱서 (int + location)
    - ix  : 지금버전에선 노추천  loc + iloc인데 중복 때문에 노추천
    - at : 스칼라(하나)값 처리
    - iat : 스칼라 처리

### loc

```python
df = pd.DataFrame(np.arange(10 ,22).reshape(3, 4), 
                  index = ["a", "b", "c"], columns = ["A", "B", "C", "D"]) # 2차원

df
```

```python
# 행우선 접근  (dataframe은 원래 열 우선 접근이다.)

print(df.loc["a"])  # 행 접근이 가능하다.
print(df[:"a"])  # 기본 인덱스  (행 접근 하려면 : 해줬어야만 함)
print("------------")
print(df.loc["a":"b"])  #행 범위도 가능
print("------------")
print(df["a":"b"])

df.A > 15     # A열에 있는 값이 15보다 큰가? 

df.loc[df.A>15]  # 그 기준에 맞는 행만 뽑아 냄

def select_rows(df, num) :
    return df.A > num


select_rows(df, 15)
df.loc[select_rows(df, 15)]
```

```python
#인덱싱을 행과 열로 모두 받을 경우

df["A"]["b"]     #기본  열, 행
df.loc["b"]["A"] #loc, 전문 인덱서 사용시  행, 열

# df["A", "b"]  # 안됨
df.loc["b", "A"]  # 이건 됌, 제약에 대해 자유스럽다.  행, 열 같은 곳에 사용 가능

df.loc["b": , "A"] # b행 부터 전체행, "A"만,   가능

df.loc["a", :]  #a행, 전체열

df.loc[["a", "b"], ["B", "C"]] # a,b행  B,C열


#모든 행에 대해서 첫번째 행에 있는 값이 11보다 작거나 같은 행의 컬럼을 추출
df.loc[:, df.loc["a", :]<=11]
```

### iloc

```python
# df.loc[0,1]  #loc는 int 인덱스 접근이 안된다.
df.iloc[0,1]

# df.iloc["a", "B"]  #안됌
df.iloc[:2, 2]
df.iloc[0, -2:]   # 0번째 행 뒤에서부터 


##########
#    B  C
# c 19 20  출력
##########

df.iloc[2, 1:3]
```

### at

```python
df.loc["a", "A"]   #값을 하나만 뽑을 때 굳이 loc 쓰지말고

df.at["a", "A"]    # at를 쓰자,  성능적으로 데이터량이 엄청 많을 때 한가지 데이터 뽑기에 at가 유리함
```

## apply

    - 행이나 열 단위로 더 복잡한 처리를 하고자 할때 사용
    - 인수로 행 또는 열을 받는 함수를 apply()의 인수로 넣으면 각 열(또는 행)을 반복하여 그 함수에 적용

```python
df = pd.DataFrame({
    "A" : [1,3,4,3,4],
    "B" : [2,3,1,2,3],
    "C" : [1,5,2,4,4]
})

print(df)
```

```python
def diff(x) :
    return x.max() - x.min()

df.apply(diff)    # apply 인자값은 함수의 주소이다.   즉 df데이터를 가지고 diff를 실행해라 
#한 열씩 실행 

df.apply(diff, axis = 1)  # 이렇게 축 방향을 인자로 줄 수도 있다.
```

```python
# 각 붓꽃별로 가장 큰값과 가장 작은 값의 비율

def peak_to_peak_ratio(x) :
    return x.max() / x.min()

print(iris.groupby(iris.species).agg(peak_to_peak_ratio))      
#species로 그룹을 묶겠다. 
#agg(함수명) : 대신 함수를 호출해서 반복해주는

print("-"*40)

print(iris.groupby(iris.species).apply(peak_to_peak_ratio))     #agg랑 결과가 같네?
```

### agg(), apply()  차이점
    적용대상에 차이가.
    - agg : 모든 행과 열에 대한 각각의 값이 적용
    - apply : 어떤 특정한 행이나 열에서 적용
    
    그래서 위위 함수는 모든 행에서 실행되는 함수이기 때문에 실행 가능
    밑에가 안됐던 이유는 어떤 행이나 열에서 적용한 정렬이기 때문에 apply만 됐던 것.
    
    
    보통 agg를 더 선호하는데 agg에는 특별한 기능이 있기 때문이다.
    
    agg는 함수를 여러개 동시에 적용가능하기 때문
    apply는 한번에 하나만 적용

```python
iris.groupby("species").agg([np.sum, np.mean, np.std])     # 여러개 적용 가능
```

### pivot_table()

    짱짱임. 근데 왜 이것만 쓰면 되는데 안쓰냐?  간단한 것은 pivot만 쓰는게 나음
    DataFrame.pivot_table(self, values=None, index=None, columns=None, aggfunc='mean', fill_value=None,
                          margins=False, dropna=True, margins_name='All', observed=False)
                          
      - data : 분석할 데이터 프레임
      - values : 분석할 데이터프레임의 분석할 열
      - aggfunc : 분석 메소드 
      - fill_value : NaN대신 대체할 값
      - dropna : NaN에 해당하는 값을 버릴지 여부
      - margins : 행과 열 끝에 소계 추가
     

```python
data = {
    "도시": ["서울", "서울", "서울", "부산", "부산", "부산", "인천", "인천"],
    "연도": ["2015", "2010", "2005", "2015", "2010", "2005", "2015", "2010"],
    "인구": [9904312, 9631482, 9762546, 3448737, 3393191, 3512547, 2890451, 263203],
    "지역": ["수도권", "수도권", "수도권", "경상권", "경상권", "경상권", "수도권", "수도권"]
}
df = pd.DataFrame(data)
df
```

```python
#pivot
df.pivot("도시", "연도", "인구")  #도시별로 묶고 그다음 연도별로 인구를 출력

#pivot table
df.pivot_table("인구", "도시", "연도")  #값, 인덱스, 컬럼 순
pd.pivot_table(df, "인구", "도시", "연도")  # df, 값, 인덱스, 컬럼
```

```python
df.pivot_table("인구", "도시", "연도", aggfunc = "sum")  
#.agg() 하던걸 이렇게 쓴다. 굳이 여기까지 쓸거면 pivot_table을 쓸 필욘 없다.


df.pivot_table("인구", "도시", "연도", aggfunc = "sum", margins=True, margins_name="합계")  
#소계를 자동으로 뽑아준다. 이게 pivot_table을 쓰는 이유, margins_name은 컬럼 이름 붙이는것 안붙이면 ALL

df.pivot_table("인구", "도시", "연도", aggfunc = "sum", margins=True, margins_name="합계",
              fill_value = 0)    # fill_value 결측치가 있다면 0으로


```

```python
# sample  유명한것    Tip데이터 : 식사 대금 대비 팁의 비율이 어떤 경우에 가장 높아지는지 찾는 데이터

#size 손님수 
import seaborn as sns
tips = sns.load_dataset("tips")
tips.head()
tips.tail()
```

```python
# 식사 대금과 팁의 비율(tip/total_bill) 을 나타내는 tip_pct를 추가   다음 시험 문제
print("팁 비율")
tips["tip_pct"] = (tips["tip"]/tips["total_bill"])
print(tips)
print("-"*80)

# 성별로 인원수 파악
print("성별 인원수")
sex_cnt = tips.pivot_table(index=None, columns="sex", values="size", aggfunc="sum").T
print(sex_cnt)
print("-"*80)
# 성별로 흡연유무별로 인원수 파악
print("성별 흡연 유무 인원수")
smoker_cnt = tips.pivot_table(index="sex", columns="smoker", values="size", aggfunc="sum").T
print(smoker_cnt)
print("-"*80)

# 위의 두 정보를 하나의 테이블로 출력(pivot_table)  1, 3번
print("두 정보 하나의 테이블")
result = pd.pivot_table(tips, index=["sex", "smoker"], values=["size", "tip_pct"], aggfunc ={"size":np.sum,"tip_pct":np.mean} ) 
print(result)
print("-"*80)

# 성별에 따른 평균 팁 비율
print("성별에 따른 평균팁 비율")
print(pd.pivot_table(tips, index="sex", values="tip_pct", aggfunc="mean"))
print("-"*80)

# 흡연 여부에 따른 팁 비율
print("흡연 여부에 따른 평균팁 비율")
print(pd.pivot_table(tips, index="smoker", values="tip_pct", aggfunc="mean"))
print("-"*80)


# 위의 정보를 하나의 테이블로 구현
print("성별, 흡연여부에 따른 평균팁 비율")
result = pd.pivot_table(tips, index=["sex", "smoker"], values="tip_pct", aggfunc ="mean") 
print(result)
print("-"*80)

# 팁의 비율이 요일과 점심/저녁 여부, 인원수에 어떤 영향을 받았는가
print("요일/점심저녁/인원수에 따른 비율")
result = pd.pivot_table(tips, index=["day", "time","size"], values="tip_pct", aggfunc ="mean") 
print(result)
print("-"*80)

# 성별, 흡연 유무별로 가장 많은 팁과 가장 적은 팁의 차이
print("성별, 흡연유무별 많은팁과 적은팁 차이")
result = pd.pivot_table(tips, index=["sex", "smoker"], values="tip", aggfunc =[max, min]) 
result["차이"] = result["max"]-result["min"]
print(result)
```

```python
#강사님.... 문제가 왜 달라
tips["tip_pct"] = tips["tip"] /tips["total_bill"]
print(tips)

# 성별 인원수
print(tips.groupby("sex").count())
# print(tips.groupby("sex").size())  #두개 같음

#성별 흡연별
print(tips.groupby(["sex","smoker"]).size())


#두정보 합치기
print(tips.pivot_table(values="tip_pct", index="sex", columns="smoker", aggfunc="count", margins=True))

#성별에 따른 평균 팁 비율
print(tips.groupby("sex")[["tip_pct"]].mean())

#흡연비율에 따른 
print(tips.groupby("smoker")[["tip_pct"]].mean())

#두정보 합치기
tips.pivot_table("tip_pct", "sex", "smoker")
tips.pivot_table("tip_pct", ["sex", "smoker"])  #관점의 차이, 결과는 같음, 테이블 모양만 다름

# 팁의 비유이 요일, 점심/저녁, 인원에 영향이 있냐
tips.pivot_table("tip_pct", "day")
tips.pivot_table("tip_pct", "time")
tips.pivot_table("tip_pct", "size")

tips.pivot_table("tip_pct", ["day", "time", "size"])


###############################################
# groupby로 해결
tips.groupby(["day", "time", "size"])[["tip_pct"]].mean()  #pivot과 차이점은 결측치 처리가 안되서 출력
tips.groupby(["day", "time", "size"])[["tip_pct"]].mean().dropna()  # 결측치 제거 함수를 써야한다.


# 성별, 흡연별 가장 많은 팁과 적은팁 차이
def peak_to_peak(x):
    return x.max() - x.min()

tips.groupby(["sex", "smoker"])[["tip"]].apply(peak_to_peak)
tips.groupby(["sex", "smoker"]).agg({"tip_pct":peak_to_peak, "total_bill":peak_to_peak})
```

```python
#범위 지정
pd.date_range("2020-1-1", "2020-4-30")
pd.date_range("2020-4-1", periods = 30)  # periods로 범위를 준다 30개, 결과에 freq=D (Day) 단위를 뜻함

# """
# freq
#     s : 초
#     T : 분
#     H : 시간
#     D : 일
#     B : 주말이 아닌 평일
#     W : 주(일요일)
#     W-MON : 주(월요일)
#     M : 각 달의 마지막 날
#     MS : 각 달의 첫 날
#     BM : 주말이 아닌 평일중에서 각 달의 마지막 날
#     BMS : 주말이 아닌 평일 중에서 각 달의 첫날
#     WOM-2THU : 각 달의 두번째 목요일
#     Q-JAN : 각 분기의 첫 달의 마지막 날
#     Q-DEC : 각 분기의 마지막 달의 마지막 날
# """
```

```python
#범위 지정 옵션
pd.date_range("2020-1-1", "2020-1-31", freq = "B") #B는 평일만 계산한다.
```

```python
#shift 연산

np.random.seed(0)

ts = pd.Series(np.random.randn(4), index=pd.date_range("2020-1-1", periods=4, freq = "M"))
print(ts)

print("-"*80)

print(ts.shift(1))  #value가 한칸씩 이동한 것.
print("-"*80)
print(ts.shift(-1)) #value가 반대로 한칸 이동

print("-"*80)
print(ts.shift(1, freq="M")) # 월이 한칸 이동함 


print("-"*80)
print(ts.shift(1, freq="W")) # 다음 일요일로 한칸 이동함 


```

```python
# resampling  : 재조정,   시간간격을 다시 조정하는 것
"""
    up-sampling : 시간 구간이 작아지면서 데이터양은 증가,, 간격이 작아지기때문
    down-sampling : 구간이 커지면서 데이터양은 감소
"""


ts = pd.Series(np.random.randn(100), index=pd.date_range("2020-1-1", periods = 100, freq ="D"))
ts.tail(10)
```

```python
#간격을 조정하므로써 데이터 갯수가 바뀐다.  100개(하루단위) -> 15개(일주일단위)
ts.resample("W").mean()

ts.resample("M").mean()        # 15개(일주일) -> 4개(월)
```

```python
# 시고저종 ?
ts = pd.Series(np.random.randn(60), index = pd.date_range("2020-1-1", periods=60, freq="T"))
ts.resample("5T").ohlc()  #5분간격  open, high, low, close
```

```python
#up-sampling
"""
    forward filling : 앞에서 나온 데이터를 그대로 사용
    backward filling : 뒤에서 나올 데이터를 앞에서 미리 사용
"""

ts.resample('30s').ffill().head(20)  #f filling
ts.resample('30s').bfill().head(20)  #b filling

```
